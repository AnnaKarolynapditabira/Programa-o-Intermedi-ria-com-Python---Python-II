#### Q3. 
Crie um dataframe `participacao` com as colunas `país`, `copa`, e `num_copas` que armazena a quantidade de copas do mundo que cada país participou, informando se é da copa masculina ou feminina. Ao final imprima, como apresentado a seguir, o top 5 países de cada competição que mais participou de copas do mundo.

Exemplo de saída (valores inventados):
```
país            copa        num_copas
Brazil          Feminina    8
Unites States   Feminina    8
Germany         Feminina    8
Japan           Feminina    7

```

```
país            copa        num_copas
Brazil          Masculina   21
```

````python
# Une os países que jogaram como time_1 e time_2 em uma única coluna 'país'
paises_time_1 = wc[['ano', 'copa', 'time_1']].rename(columns={'time_1': 'país'})
paises_time_2 = wc[['ano', 'copa', 'time_2']].rename(columns={'time_2': 'país'})

# Junta todos os jogos em que o país participou
todos_paises = pd.concat([paises_time_1, paises_time_2])

# Remove duplicatas para contar uma vez por ano
participacao_unica = todos_paises.drop_duplicates()

# Conta o número de participações por país e tipo de copa
participacao = participacao_unica.groupby(['país', 'copa']).size().reset_index(name='num_copas')

# Mostra o top 5 Feminina
print(participacao[participacao['copa'] == 'Feminina'].sort_values(by='num_copas', ascending=False).head(5))

# Linha separadora
print("\n")

# Mostra o top 5 Masculina
print(participacao[participacao['copa'] == 'Masculina'].sort_values(by='num_copas', ascending=False).head(5))
````

#### Q4. 
* Crie um dataframe `gols` com duas colunas `país` e `total_gols` com o total de gols marcados por cada país durante todas as copas do mundo, juntando gols em casa (`gols_1`) e gols como time visitante (`gols_2`).
* Imprima o dataframe `gols` ordenado de forma decrescente, para que os times com mais gols fiquem no topo.

Segue um exemplo ilustrativo com o formato do dataframe resultado:

```
país        total_gols
Brazil      120
Argentina   112
Germany     107
...
```

````python
import pandas as pd

# Supondo que o DataFrame wc já esteja carregado e com os nomes traduzidos

# Soma os gols feitos como time da casa (gols_1)
gols_mandante = wc.groupby('time_1')['gols_1'].sum().reset_index()
gols_mandante.columns = ['país', 'gols_casa']

# Soma os gols feitos como visitante (gols_2)
gols_visitante = wc.groupby('time_2')['gols_2'].sum().reset_index()
gols_visitante.columns = ['país', 'gols_fora']

# Junta os dois DataFrames
gols_total = pd.merge(gols_mandante, gols_visitante, on='país', how='outer').fillna(0)

# Calcula o total de gols
gols_total['total_gols'] = gols_total['gols_casa'] + gols_total['gols_fora']

# Ordena do maior para o menor
gols_ordenado = gols_total[['país', 'total_gols']].sort_values(by='total_gols', ascending=False).reset_index(drop=True)

# Exibe o resultado
print(gols_ordenado.head(10))  # Top 10 países com mais gols
````

#### Q5. 
Qual país tomou mais cartões amarelos somando todas as copas?

Neste exercício você vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observação é uma string represetando uma lista dos cartões amarelos de um único jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16’|1:0|Rosana Gómez', '20’|2:0|Gabriela Chávez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o método genérico `apply` para chamar uma função que você vai criar para transformar uma observação de cartão amarelo em uma contagem de elementos da lista. 

Lembre de levar em consideração que muitas observações são `NaN`. 

Em seguida faça sua mágica para agrupar as informações por país, acumular os cartões de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
país        cartões amarelos
Argentina   97
England     93
Australia   93
...
```

````python
import pandas as pd
import numpy as np

# Função que conta cartões amarelos a partir da string
def contar_cartoes(valor):
    if pd.isna(valor):
        return 0
    # Remove colchetes e aspas caso esteja formatado como string de lista
    valor = valor.strip("[]").replace("'", "").replace('"', '')
    if valor == '':
        return 0
    return len(valor.split(',')) if ',' in valor else len(valor.split('|')) if '|' in valor else 1

# Aplica a função para contar os cartões
wc['num_cartoes_1'] = wc['cartao_amarelo_1'].apply(contar_cartoes)
wc['num_cartoes_2'] = wc['cartao_amarelo_2'].apply(contar_cartoes)

# Soma os cartões recebidos em casa
cartoes_casa = wc.groupby('time_1')['num_cartoes_1'].sum().reset_index()
cartoes_casa.columns = ['país', 'cartoes']

# Soma os cartões recebidos como visitante
cartoes_fora = wc.groupby('time_2')['num_cartoes_2'].sum().reset_index()
cartoes_fora.columns = ['país', 'cartoes']

# Junta as duas contagens
cartoes_total = pd.concat([cartoes_casa, cartoes_fora])
cartoes_total = cartoes_total.groupby('país')['cartoes'].sum().reset_index()

# Ordena do maior para o menor
cartoes_total = cartoes_total.sort_values(by='cartoes', ascending=False).reset_index(drop=True)

# Exibe os 10 países que mais receberam cartões amarelos
print(cartoes_total.head(10))
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de pênalti.

Para conseguir essa informação, você precisará trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas também são colunas textuais, onde cada observações apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol está na forma `'Jogador(a) · minuto’'`. Por exemplo:
```
'Alex Morgan · 12’|Rose Lavelle · 20’'
```

Lembre de levar em consideração que muitas observações são `NaN`. 

Recomendo criar um dicionário à parte, onde cada chave será um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente será a contagem de ocorrências desses nomes.

Em seguida basta converter o seu dicionário em um dataframe e imprimí-lo na forma (valores não são inventados 😁):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
import pandas as pd
from collections import defaultdict

# Função para extrair os nomes dos jogadores da string de gols
def extrair_jogadores(gol_str):
    if pd.isna(gol_str):
        return []
    jogadores = []
    eventos = gol_str.strip().split('|')
    for evento in eventos:
        if '·' in evento:
            jogador = evento.split('·')[0].strip()
            jogadores.append(jogador)
    return jogadores

# Inicializa um dicionário para contar gols por jogador
contagem_gols = defaultdict(int)

# Lista de colunas com informações de gols
colunas_gols = ['gols_1_detalhes', 'gols_2_detalhes', 'gols_1_penalti', 'gols_2_penalti']

# Conta os gols de cada jogador
for coluna in colunas_gols:
    wc[coluna].dropna().apply(lambda x: [contagem_gols[jogador] := contagem_gols[jogador] + 1 for jogador in extrair_jogadores(x)])

# Converte o dicionário em DataFrame
df_goleadores = pd.DataFrame(list(contagem_gols.items()), columns=['jogador(a)', 'num_gols'])

# Ordena pelo número de gols
df_goleadores = df_goleadores.sort_values(by='num_gols', ascending=False).reset_index(drop=True)

# Exibe os 10 maiores goleadores
print(df_goleadores.head(10))
````
